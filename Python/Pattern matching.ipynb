# Using Python's `match` Statement (Structural Pattern Matching)

Introduced in Python 3.10, the `match` statement (often called structural pattern matching) provides a powerful way to handle complex conditional logic more cleanly than long `if-elif-else` chains. Here's a comprehensive guide on how and when to use it:

## When to Use `match`

### Best Use Cases:
1. **Replacing long if-elif chains** - Especially when checking the same value against multiple patterns
2. **Handling different data structures** - When processing lists, tuples, or dictionaries with specific shapes
3. **Parsing protocols/commands** - Like in network protocols or CLI commands (as in your example)
4. **Processing ASTs or complex nested data** - Common in compilers, interpreters, and data transformers
5. **Implementing state machines** - Where different states require different handling

### When NOT to Use:
- Simple conditions that can be expressed with a single `if`
- Type checking (use `isinstance()` instead)
- When you need backward compatibility with Python <3.10

## Basic Syntax

```python
match subject:
    case pattern1:
        # handle pattern1
    case pattern2:
        # handle pattern2
    case _:
        # default case
```

## Common Pattern Types

### 1. Literal Patterns
```python
match status:
    case 200:
        print("Success")
    case 404:
        print("Not found")
    case 500:
        print("Server error")
```

### 2. Sequence Patterns (Lists/Tuples)
```python
match command.split():
    case ["go", direction]:
        move(direction)
    case ["attack", target]:
        attack(target)
    case ["quit"]:
        exit_game()
```

### 3. Mapping Patterns (Dictionaries)
```python
match config:
    case {"mode": "production", "port": port}:
        start_server(port, debug=False)
    case {"mode": "development", "port": port}:
        start_server(port, debug=True)
```

### 4. Class Patterns
```python
match event:
    case Click(x, y):
        handle_click(x, y)
    case KeyPress(key="Enter"):
        handle_enter()
    case KeyPress(key):
        handle_key(key)
```

### 5. OR Patterns
```python
match value:
    case 0 | 1 | 2:
        print("Small number")
    case str() | bytes():
        print("String-like")
```

### 6. Capture Patterns (Binding Variables)
```python
match points:
    case (x, y):
        print(f"X: {x}, Y: {y}")
    case (x, y, z):
        print(f"3D point: {x},{y},{z}")
```

### 7. Wildcard Pattern
```python
match data:
    case []:
        print("Empty list")
    case [_]:
        print("Single element list")
    case [_, _]:
        print("Two elements")
    case _:
        print("Something else")
```

## Advanced Features

### Guards (Additional Conditions)
```python
match point:
    case (x, y) if x == y:
        print("On the diagonal")
    case (x, y):
        print("Somewhere else")
```

### Nested Patterns
```python
match response:
    case {"status": 200, "data": [{"name": name}, *_]}:
        print(f"First item: {name}")
    case {"status": 404}:
        print("Not found")
```

## Practical Example: Command Processor

```python
def handle_command(command):
    match command.split():
        case ["load", filename]:
            load_file(filename)
        case ["save", filename]:
            save_file(filename)
        case ["exit" | "quit"]:
            shutdown()
        case ["help"]:
            print_help()
        case _:
            print(f"Unknown command: {command}")
```

## Performance Considerations

- `match` is generally as fast as equivalent `if-elif` chains
- The interpreter checks patterns in order, so put the most common cases first
- Complex patterns with many guards will be slower

## Key Benefits

1. **Readability** - Clearly shows the relationship between patterns and actions
2. **Safety** - Exhaustiveness checking (with `_` catch-all) prevents unhandled cases
3. **Expressiveness** - Can match against structure, not just values
4. **Maintainability** - Easier to add new cases than long if-elif chains

Remember that while `match` is powerful, it's not always the right tool - use it when it genuinely makes your code clearer and more maintainable.

In Python, the `array` module provides an `array()` function to create arrays that are more efficient than lists for storing homogeneous data types. Here's how to use it:

### 1. **Import the `array` module**
```python
from array import array
```

### 2. **Create an array**
The `array()` constructor takes two arguments:
- **Type code**: A string specifying the data type of the array.
- **Initializer (optional)**: A list or iterable to initialize the array.

#### **Common Type Codes**
| Type Code | C Type      | Python Type | Size (bytes) |
|-----------|-------------|-------------|--------------|
| `'b'`     | signed char | int         | 1            |
| `'B'`     | unsigned char | int       | 1            |
| `'i'`     | signed int  | int         | 2 or 4       |
| `'I'`     | unsigned int | int        | 2 or 4       |
| `'f'`     | float       | float       | 4            |
| `'d'`     | double      | float       | 8            |

#### **Example: Creating an array**
```python
# Array of integers (signed)
arr_int = array('i', [1, 2, 3, 4, 5])

# Array of floats
arr_float = array('f', [1.5, 2.5, 3.5])

# Array of unsigned integers
arr_uint = array('I', [10, 20, 30])
```

### 3. **Basic Array Operations**
```python
# Access elements (like lists)
print(arr_int[0])  # Output: 1

# Modify elements
arr_int[1] = 99

# Append a new element
arr_int.append(6)

# Extend with another iterable
arr_int.extend([7, 8, 9])

# Get array length
print(len(arr_int))  # Output: 8

# Convert to list
list_from_array = arr_int.tolist()
```

### 4. **Array Methods**
- `append(x)` – Add an item to the end.
- `extend(iterable)` – Append items from an iterable.
- `insert(i, x)` – Insert an item at a given position.
- `remove(x)` – Remove the first occurrence of `x`.
- `pop([i])` – Remove and return the item at position `i`.
- `index(x)` – Return the index of the first occurrence of `x`.
- `count(x)` – Count occurrences of `x`.
- `reverse()` – Reverse the array.
- `tobytes()` – Convert to a bytes object.
- `frombytes(bytes)` – Append from a bytes object.

### 5. **Example: Full Usage**
```python
from array import array

# Create an array of integers
numbers = array('i', [10, 20, 30, 40, 50])

# Modify an element
numbers[2] = 99

# Append a new number
numbers.append(60)

# Remove an element
numbers.remove(40)

# Iterate over the array
for num in numbers:
    print(num)

# Output: 10, 20, 99, 50, 60
```

### **When to Use `array` Instead of `list`?**
- When you need **memory-efficient storage** of large datasets.
- When working with **numerical data** in a constrained type (like C-style arrays).
- When interfacing with **low-level C libraries** (since `array` is compatible with C types).

For more advanced numerical operations, consider using **`numpy` arrays** instead. 🚀

Yes, Python **lists** have similar methods to **arrays** (from the `array` module), but with some differences due to their flexibility (lists can store **heterogeneous** data, while arrays store **homogeneous** data). Below is a comparison:

---

### **1. Similar Methods Between `list` and `array`**
| Method         | `list` (Built-in) | `array` (`array` module) | Description |
|---------------|------------------|------------------|-------------|
| `append(x)` | ✅ Yes | ✅ Yes | Adds `x` to the end. |
| `extend(iterable)` | ✅ Yes | ✅ Yes | Extends with elements from an iterable. |
| `insert(i, x)` | ✅ Yes | ✅ Yes | Inserts `x` at index `i`. |
| `remove(x)` | ✅ Yes | ✅ Yes | Removes the first occurrence of `x`. |
| `pop([i])` | ✅ Yes | ✅ Yes | Removes and returns the item at index `i`. |
| `index(x)` | ✅ Yes | ✅ Yes | Returns the first index of `x`. |
| `count(x)` | ✅ Yes | ✅ Yes | Counts occurrences of `x`. |
| `reverse()` | ✅ Yes | ✅ Yes | Reverses the sequence in place. |
| `copy()` | ✅ Yes | ❌ No (but can use `array('typecode', arr)`) | Returns a shallow copy. |
| `clear()` | ✅ Yes | ❌ No (but can reassign `arr = array('typecode')`) | Removes all elements. |

---

### **2. Key Differences**
| Feature | `list` | `array` |
|---------|--------|---------|
| **Data Type** | Can store **any** Python objects (heterogeneous). | Only stores **homogeneous** elements (specified by type code). |
| **Memory Efficiency** | Less efficient for large numerical data. | More efficient (like C arrays). |
| **Performance** | Slower for numerical operations. | Faster for numerical operations (due to fixed type). |
| **Type Flexibility** | Can mix `int`, `str`, `float`, etc. | Must be uniform (e.g., all `int` or all `float`). |
| **Additional Methods** | Supports `sort()`, `copy()`, `clear()`. | Supports `tobytes()`, `frombytes()`, `fromlist()`. |

---

### **3. Example Comparison**
#### **List Example**
```python
my_list = [1, 2, 3, 4, 5]

my_list.append(6)          # [1, 2, 3, 4, 5, 6]
my_list.extend([7, 8])     # [1, 2, 3, 4, 5, 6, 7, 8]
my_list.remove(3)          # [1, 2, 4, 5, 6, 7, 8]
popped = my_list.pop(1)    # popped=2 → [1, 4, 5, 6, 7, 8]
idx = my_list.index(5)     # idx=2
count = my_list.count(4)   # count=1
my_list.reverse()          # [8, 7, 6, 5, 4, 1]
```

#### **Array Example**
```python
from array import array
my_array = array('i', [1, 2, 3, 4, 5])

my_array.append(6)          # array('i', [1, 2, 3, 4, 5, 6])
my_array.extend([7, 8])     # array('i', [1, 2, 3, 4, 5, 6, 7, 8])
my_array.remove(3)          # array('i', [1, 2, 4, 5, 6, 7, 8])
popped = my_array.pop(1)    # popped=2 → array('i', [1, 4, 5, 6, 7, 8])
idx = my_array.index(5)     # idx=2
count = my_array.count(4)   # count=1
my_array.reverse()          # array('i', [8, 7, 6, 5, 4, 1])
```

---

### **4. When to Use Which?**
- **Use `list` when:**
  - You need mixed data types (`int`, `str`, `float`, etc.).
  - You need built-in methods like `sort()`, `copy()`, `clear()`.
  - Memory efficiency is not a major concern.

- **Use `array` when:**
  - You are working with **large numerical datasets** (better memory efficiency).
  - You need **fixed-type** storage (like C arrays).
  - You need **faster numerical operations** (but `numpy` is even better for this).

For **advanced numerical computing**, consider **NumPy arrays** (`numpy.ndarray`), which offer even better performance and functionality. 🚀
